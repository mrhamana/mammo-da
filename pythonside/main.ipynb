```python
# Calculate different types of skewness for comparison
print("Regular Skewness:", skewness(radius_mean[1]))
print("Pearson Skewness:", pearson_skewness(radius_mean[1]))
print("Bowley Skewness:", bowley_skewness(radius_mean[1]))

# Calculate variation statistics
print("\nVariation Statistics:")
print("Coefficient of SD:", coefficient_sd(radius_mean[1]))
print("Coefficient of Variation (%):", coefficient_variation(radius_mean[1]))
print("IQR:", iqr(radius_mean[1]))
print("Quartile Deviation:", quartile_deviation(radius_mean[1]))
```

```python
# Correlation Analysis
print("Covariance:", covariance(radius_mean[1], texture_mean[1]))
print("Regression Slope:", regression_slope(radius_mean[1], texture_mean[1]))
print("R-squared:", r_squared(radius_mean[1], texture_mean[1]))
```

```python
# Regression analysis between radius and texture
slope = regression_slope(radius_mean[1], texture_mean[1])
r2 = r_squared(radius_mean[1], texture_mean[1])
print("Regression slope:", slope)
print("R-squared value:", r2)
```

```python

# Basic Statistical Measures
print("Midrange:", midrange(radius_mean[1]))
print("IQR:", iqr(radius_mean[1]))
print("Quartile Deviation:", quartile_deviation(radius_mean[1]))
print("Coefficient of SD:", coefficient_sd(radius_mean[1]))
print("Coefficient of Variation (%):", coefficient_variation(radius_mean[1]))

# Advanced Statistical Measures
print("Raw Moments (3rd order):", raw_moments(3, radius_mean[1]))
print("Pearson Skewness:", pearson_skewness(radius_mean[1]))
print("Bowley Skewness:", bowley_skewness(radius_mean[1]))
```
