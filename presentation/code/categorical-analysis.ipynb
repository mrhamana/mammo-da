```python
import sys
print(sys.executable)
sys.path.append('../../pythonside')

import extract_data as ed
from functions import mode
import graph as gh # Import graph for potential visualization later
import matplotlib.pyplot as plt # Import matplotlib for potential visualization later
```

```python
file = '../../data/breast-country.csv'
column_name, cancer_labels = ed.extractinfo(file, [2, 'Cancer label']) # Assuming 'Cancer label' is in the 2nd column (index 1)
```

```python
# Calculate the mode
cancer_label_mode = mode(cancer_labels)

# Count the occurrences of each category for potential visualization or further analysis
label_counts = {}
for label in cancer_labels:
    label_counts[label] = label_counts.get(label, 0) + 1

# Print the mode and label counts
print("=== CATEGORICAL DATA ANALYSIS (Cancer Label) ===")
print(f"Mode Cancer Label: {cancer_label_mode}")
print("\nLabel Counts:")
for label, count in sorted(label_counts.items(), key=lambda item: item[1], reverse=True): # Sort by count descending
    print(f"- {label}: {count}")


# Optional: Basic bar chart visualization (if needed later)
labels = list(label_counts.keys())
counts = list(label_counts.values())
plt.figure(figsize=(10, 6))
plt.bar(labels, counts, color='skyblue')
plt.xlabel('Cancer Label')
plt.ylabel('Frequency')
plt.title('Distribution of Cancer Labels')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
