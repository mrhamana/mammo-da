```python
# Cell 1: Imports and Data Setup
import sys
sys.path.append('../../pythonside')

import extract_data as ed
import numpy as np
import matplotlib.pyplot as plt
from functions import normal_pdf  # We'll use this for PDF estimation
import graph as gh

# Load the dataset
file = '../../data/breast-country.csv'
column_name, asr_data = ed.extractinfo(file, [7, 'ASR (World) per 100 000'])
```

```python
# Cell 2: Histogram Visualization
gh.estimate_pdf(
    data=asr_data,
    num_points=1000,  # Moved up as 2nd required parameter
    title=f'Probability Density Estimation: {column_name}',
    xlabel='Cases per 100,000',
    ylabel='Density',
    figsize=(10, 6)
)
plt.show()
```

```python
# Cell 3: Normal Distribution Plot
gh.normal_graph(
    data=asr_data,
    std_dev_range=3,  # Show 3 standard deviations from mean
    title=f'Normal Distribution Fit: {column_name}',
    figsize=(10, 6)
)
plt.show()
```

```python
# Cell 4: Q-Q Plot
gh.qq_plot(
    data=asr_data,
    title=f'Q-Q Plot: {column_name}',
    xlabel='Theoretical Quantiles',  
    ylabel='Sample Quantiles',      
    figsize=(8, 8)
)
```

```python
# Cell 5: PDF Estimation
gh.estimate_pdf(
    data=asr_data,
    title=f'Probability Density Estimation: {column_name}',
    xlabel='Cases per 100,000',
    ylabel='Density',
    figsize=(10, 6),
    num_points=1000  # Resolution of the PDF curve
)
plt.show()
```
